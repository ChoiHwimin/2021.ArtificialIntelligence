```python
import torch
torch.manual_seed(1)
```

```python
train=pd.read_csv('/kaggle/input/sejonguniai-12312312/TrainData.csv')
test=pd.read_csv('/kaggle/input/sejonguniai-12312312/TestData.csv')
submit=pd.read_csv('/kaggle/input/sejonguniai-12312312/submit.csv')
train
```

```python
train['최대 순간 풍속 풍향(16방위)']=np.sin(train['최대 순간 풍속 풍향(16방위)'].values)
train['최대 풍속 풍향(16방위)']=np.sin(train['최대 풍속 풍향(16방위)'].values)
test['최대 순간 풍속 풍향(16방위)']=np.sin(test['최대 순간 풍속 풍향(16방위)'].values)
test['최대 풍속 풍향(16방위)']=np.sin(test['최대 풍속 풍향(16방위)'].values)
```

```python
xtrain=train.drop(['ID','호기','평균'],axis=1)
ytrain=train['평균']
xtest=test.drop(['ID','호기'],axis=1)
```

```python
xtrain
```

```python
xtest
```

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
scxtrain=scaler.fit_transform(xtrain)
scxtest=scaler.transform(xtest)
```

```python
x=torch.FloatTensor(scxtrain)
y=torch.FloatTensor(ytrain.values)
xt=torch.FloatTensor(scxtest)
```

```python
linear1=torch.nn.Linear(x.shape[1],32)
linear2=torch.nn.Linear(32,64)
linear3=torch.nn.Linear(64,32)
linear4=torch.nn.Linear(32,1)
relu=torch.nn.ReLU()
dropout=torch.nn.Dropout(p=0.3)
```

```python
torch.nn.init.xavier_normal_(linear1.weight)
torch.nn.init.xavier_normal_(linear2.weight)
torch.nn.init.xavier_normal_(linear3.weight)
torch.nn.init.xavier_normal_(linear4.weight)
```

```python
model=torch.nn.Sequential(
    linear1,relu,dropout,
    linear2,relu,dropout,
    linear3,relu,dropout,
    linear4)
```

```python
loss=torch.nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.0001)
```

```python
model.train()

for epoch in range(37900):
    model.train()
    optimizer.zero_grad()
    cost=loss(model(x),y.unsqueeze(1))
    cost.backward()
    optimizer.step()
    if epoch % 2000 == 0:
        print('epoch :',epoch,'cost {:.2f}'.format(cost.item()))
```

```python
model.eval()
with torch.no_grad():
    prediction=model(xt).numpy()
    print(prediction)
```

```python
submit.TARGET = prediction
submit
```

```python
submit.to_csv('submission.csv',index=False)
```
